### Code Snippet to Retreive Machine Master Table Begin/End Timestamps

```python
database_df = pd.read_csv("machine1 master table results.csv")

# Filter the rows where 'button_name' is 'Begin Main Study' to get the start timestamp
start_df = database_df[database_df['button_name'].str.contains('Begin Main Study', na=False)]

# Filter the rows where 'button_name' is 'Submit' to get the end timestamp
# We assume the last "Submit" is the end of the study for each user
end_df = database_df[database_df['button_name'].str.contains('Submit', na=False)].groupby('user_id').tail(1)

# Merge start and end timestamps based on 'user_id'
machine_master_df = pd.merge(start_df[['user_id', 'timestamp']], 
                     end_df[['user_id', 'timestamp']], 
                     on='user_id', 
                     suffixes=('_start', '_end'))

# Show the dataframe
print(machine_master_df)

# Save the result to a CSV
machine_master_df.to_csv('user_timestamps.csv', index=False)

# Display the resulting DataFrame information
machine_master_df.info()
```

### Changing the Datatype of Begin/End Timestamps to be usable

```python
# Convert the timestamps to datetime format
machine_master_df['timestamp_start'] = pd.to_datetime(machine_master_df['timestamp_start'], format='%I:%M:%S %p')
machine_master_df['timestamp_end'] = pd.to_datetime(machine_master_df['timestamp_end'], format='%I:%M:%S %p')

# Calculate duration of time
machine_master_df['duration'] = machine_master_df['timestamp_end'] - machine_master_df['timestamp_start']

# Display the updated DataFrame
print(machine_master_df)
```

### Extract Timestamps from .csv into df

```python
# Function to extract the timestamp from the column name
def extract_timestamp(df):
    # Get the column name that contains the timestamp
    timestamp_column_name = df.columns[3]  # Assuming the timestamp column is always at index 3
    # Use regex to extract the timestamp string from the column name
    timestamp_match = re.search(r'\d{4}/\d{2}/\d{2} \d{2}:\d{2}:\d{2}\.\d{3}', timestamp_column_name)
    if timestamp_match:
        # Convert to datetime
        return pd.to_datetime(timestamp_match.group(0))
    return None

# List of CSV files with user identifiers
file_paths = [
    ("C:/Users/Kelly/Desktop/Left_gaze/User 0_all_gaze.csv", 0),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 1_all_gaze.csv", 1),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 2_all_gaze.csv", 2),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 3_all_gaze.csv", 3),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 5_all_gaze.csv", 5),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 7_all_gaze.csv", 7),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 8_all_gaze.csv", 8),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 9_all_gaze.csv", 9),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 10_all_gaze.csv", 10),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 11_all_gaze.csv", 11),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 12_all_gaze.csv", 12),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 13_all_gaze.csv", 13),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 14_all_gaze.csv", 14),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 16_all_gaze.csv", 16),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 17_all_gaze.csv", 17),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 18_all_gaze.csv", 18),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 19_all_gaze.csv", 19),
    ("C:/Users/Kelly/Desktop/Left_gaze/User 20_all_gaze.csv", 20)
]

# List to store extracted timestamps
timestamps = []

# Iterate through the list of files
for file_path, user_id in file_paths:
    # Load the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Extract timestamp
    timestamp = extract_timestamp(df)
    if timestamp is not None:
        timestamps.append({"User_ID": user_id, "Extracted_Timestamp": timestamp})

# Create a new DataFrame with the extracted timestamps
timestamps_df = pd.DataFrame(timestamps)

# Display the timestamps DataFrame
print(timestamps_df)
```

### Convert Extracted timestamp to Usable value

```python
# Convert the 'Extracted_Timestamp' column to datetime format
timestamps_df['Extracted_Timestamp'] = pd.to_datetime(timestamps_df['Extracted_Timestamp'], format='%I:%M:%S %p').dt.time

# Display the updated DataFrame
print(timestamps_df)
```

```python
from datetime import datetime

# Sample DataFrame for demonstration (use your actual DataFrame)
data = {
    'timestamp_start': ['12:25:37 PM', '12:20:45 PM', '12:23:58 PM', '12:25:47 PM'],
}
merged_df = pd.DataFrame(data)

# Convert the 'timestamp_start' to datetime with correct format
merged_df['timestamp_start'] = pd.to_datetime(merged_df['timestamp_start'], format='%I:%M:%S %p')

# Example extracted time string
extracted_time_str = "12:23:38.733"  # This should be in 24-hour format for the extraction
extracted_time = pd.to_datetime(extracted_time_str, format='%H:%M:%S.%f').time()

# Combine extracted_time with the date from 'timestamp_start'
merged_df['timestamp_start_with_date'] = merged_df['timestamp_start'].apply(
    lambda x: datetime.combine(x.date(), extracted_time)
)

# Subtract the two datetime values to get the time difference
merged_df['time_difference'] = merged_df['timestamp_start'] - merged_df['timestamp_start_with_date']

# Calculate the time difference in milliseconds
merged_df['time_difference_milliseconds'] = merged_df['time_difference'].dt.total_seconds() * 1000

# Print the resulting DataFrame with time differences in milliseconds
print(merged_df[['timestamp_start', 'timestamp_start_with_date', 'time_difference', 'time_difference_milliseconds']])
```
